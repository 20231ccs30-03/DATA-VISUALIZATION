```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01"]),
    "Value":[18,20,23,27,32,36,40,43,47,53]
}
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"],marker='o')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Data-Trend Analysis")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

data={
    "Date":pd.to_datetime(["2022-01-01","2022-02-01","2022-03-01","2022-04-01","2022-05-01","2022-06-01","2022-07-01","2022-08-01","2022-09-01","2022-10-01"]),
    "Value":[9,19,8,24,37,27,43,51,39,45]
}
df=pd.DataFrame(data)
df.set_index("Date",inplace=True)
plt.figure(figsize=(10,6))
plt.plot(df.index,df["Value"])
plt.plot(df.index, df["Value"], label='Daily')
plt.legend(loc='upper left')
plt.xlabel("Date")
plt.ylabel("Value")
plt.title("Time Series Data-Trend Analysis")
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
from statsmodels.tsa.seasonal import seasonal_decompose
data = pd.read_csv('https://raw.githubusercontent.com/selva86/datasets/master/sunspotarea.csv', parse_dates=['date'], index_col='date')
result = seasonal_decompose(data, model='additive', period=12)


fig, axes = plt.subplots(1, 3, figsize=(20, 4), dpi=100)


data.plot(ax=axes[0], title='Original Data', legend=False)


result.trend.plot(ax=axes[1], title='Trend Only', legend=False)


result.seasonal.plot(ax=axes[2], title='Seasonal Only', legend=False)

plt.tight_layout()
plt.show()
```
