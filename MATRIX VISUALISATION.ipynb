```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
#heat maps
data=np.random.rand(13,16)
plt.imshow(data,cmap='magma')
plt.title('2d Heat Map')
plt.colorbar()
plt.show()
```

```python
#heat maps using seaborn library
data=np.random.randint(low=14,
                       high=100,
                       size=(3,3))
hm=sns.heatmap(data=data,annot=True)
plt.show
```
