```python
import requests
from bs4 import BeautifulSoup
import os

def extract_images(url, save_folder):
    # Make a GET request to the URL
    response = requests.get(url)

    # Check if the request was successful
    if response.status_code == 200:
        # Parse the HTML content of the page
        soup = BeautifulSoup(response.content, 'html.parser')

        # Find all image tags
        img_tags = soup.find_all('img')

        # Create a folder to save the images if it doesn't exist
        if not os.path.exists(save_folder):
            os.makedirs(save_folder)

        # Extract and save images
        for img_tag in img_tags:
            # Get the image source (URL)
            img_url = img_tag['src']

            # Get the image name
            img_name = os.path.basename(img_url)

            # Download the image
            img_data = requests.get(img_url).content

            # Save the image to the specified folder
            with open(os.path.join(save_folder, img_name), 'wb') as img_file:
                img_file.write(img_data)

        print("Images extracted and saved successfully!")
    else:
        print("Failed to fetch page content.")

# Example usage:
website_url = 'https://www.imdb.com/list/ls069457382/'
save_folder = 'images'
extract_images(website_url, save_folder)
```
